1.接口测试
    1.1.测试外部接口：只需要测试正例
    1.2.测试内部接口：测正例，异常，权限控制

2.接口测试流程
    1.拿到api（也可以用抓包）：熟悉业务，地址，鉴权方式，入参，错误码
    2.编写用例（XMIND）：
        正例：正常入参，返回数据
            请求四要素：请求方式，请求路径，请求头，请求数据
            响应四要素：响应码，响应信息，响应头，响应数据
        反例：
            权限反例：鉴权码token为空/错误/过期
            参数反例：参数为空/类型异常/长度异常
            错误码覆盖：根据业务
            其他：黑名单，调用次数，分页场景
    3.使用postman执行接口测试
    4.持续集成，输出测试报告发送邮件

3.Postman界面
    3.1.get请求:
        第一种传参Param:查询字符串参数，?后，用&分割，是url一部分
        Authorization
        Headers:请求头
        第二种传参Body:不在url，有json,表单，文件
            form-data:参数+文件
            x-www-form-urlencoded:只有参数
            raw: json,xml,html,js等等
            binary:
            GraphQL: 查询语言
        Script:前置/后置
        Cookies:属于请求头
    3.2.get响应:
        Body: 响应数据
            Pretty:以不同格式查看返回的数据
            Raw:以文本方式查看返回的数据
            Preview:以网页方式查看返回数据
        Cookies:属于响应头
        Headers: 响应头
        Test Results: 断言接口是否成功
    3.3.post请求:
        params要有access_token,每次都会变
        编辑/删除
        在body传参数


4.postman接口关联
    上一个返回值作为下一个参数,不像get用?后面传参，用body传参安全
    4.1.实现:
        4.1.1.用json对象取值，保存到全局变量
            var jsonData = pm.response.json();
            console.log(jsonData.access_token);
            pm.globals.set("saki_token",jsonData.access_token);
        4.1.2.可以使用正则提取，保存到全局变量
            var responseText = pm.response.text();
            var match = responseText.match('"access_token":"(.*?)"');
            console.log(match[1]);
            pm.globals.set("saki_token",match[1]);
        4.1.3.使用全局变量{{变量名}}
            {{saki_token}}
    4.2.如何避免脏数据，形成数据闭环:用全局
        {{$}}:postman内置的，{{}}是自己设的
    4.3.环境变量
        使用: 只能在单一环境使用
            多种环境区别：ip/域名不一样,把头改成https://{{ip}}/
    4.4.动态参数
        时间戳{{$timestamp}}
        随机数{{$randomint}}
        随机GUID字符串{{$guid}}
        自定义动态参数:var times = Date.now(); //获得时间戳
                    pm.globals.set("times",times);

9.Cookie鉴权
    第一次访问生成cookie，服务器在响应头set-cookie里面把生成的cookie信息发给客户端
    第二到N次访问，客户端在请求头cookie带上cookie
    9.1.分类
        会话cookie: 在内存，浏览器关了自动清
        持久cookie: 在硬盘，时间失效了自动清
