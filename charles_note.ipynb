1.概念
    代理服务器
    跨平台
    半免费
    工作原理：充当中间的代理服务器
2.功能
    支持Http/Https代理
    支持流量控制
    支持接口并发请求
    支持重发网络请求
    支持断点调试
3.优点
    支持不同OS
    支持按域名/接口查报文
    支持反向代理
    网络限速可选网络类型
    解析AMF协议

4.代理设置
    proxy-port:8888
5.访问控制
    proxy-Access control Settings
    
6.抓包分析问题
    前端or后端: 前端输入框没做限制，后端返回提示不清晰
7.Https
    对http数据进行了加密 -> 显示乱码 -> 安装SSL证书 + 代理配置https
8.流量配置
    Throttle Setting -> 配置弱网
9.断点配置：用来构建异常测试场景
    右键接口，breakpoints，回游览器发请求，回charles修改提交，游览器修改成功请求
